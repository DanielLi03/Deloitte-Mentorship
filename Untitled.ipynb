```python
import pandas as pd
import seaborn as sns
```

### Load tables

```python
customers = pd.read_csv("archive/customer_dim.csv", encoding = 'latin_1')
customers.head()
```

```python
facts = pd.read_csv("archive/fact_table.csv")
facts.head()
```

```python
items = pd.read_csv("archive/item_dim.csv", encoding='latin-1')
items.head()
```

```python
stores = pd.read_csv("archive/store_dim.csv")
stores.head()
```

```python
times = pd.read_csv("archive/time_dim.csv", encoding = 'latin-1')
times.head()
```

```python
transactions = pd.read_csv("archive/Trans_dim.csv", encoding = 'latin-1')
transactions.head()
```

### Basic EDA

```python
#look at total revenue over time (month to month)
sales_time = facts.merge(times, left_on='time_key', right_on='time_key', how='left')
sales_time = sales_time.groupby(['year', 'month'])['total_price'].sum().reset_index()
sales_time['total_price'].plot.line()
```

```python
# product performance
total_product = facts.groupby('item_key')['quantity'].sum().reset_index()
total_product = total_product.merge(items[['item_key', 'item_name', 'desc']], how='left', left_on='item_key', right_on='item_key').sort_values(by=['quantity'], ascending=False)
print(total_product.head(20))
total_product.tail(20)
```

```python
#revenue by district/division

store_revenue = facts[['store_key', 'total_price']].groupby('store_key').sum().reset_index()
store_revenue = store_revenue.merge(stores[['store_key', 'division', 'district']], how='left', left_on='store_key', right_on='store_key')
print(store_revenue.groupby(['division', 'district']).sum().sort_values(by=['total_price'], ascending = False))
print(store_revenue.groupby(['district']).sum().sort_values(by=['total_price'], ascending = False))
print(store_revenue.groupby(['division']).sum().sort_values(by=['total_price'], ascending = False))
```

```python
#transaction analysis

quantity_trans = facts[['payment_key', 'quantity']].groupby('payment_key').sum()
revenue_trans = facts[['payment_key', 'total_price']].groupby('payment_key').sum()
quantity_trans = quantity_trans.merge(transactions, how='left', left_on='payment_key', right_on='payment_key').groupby('trans_type').sum()
revenue_trans = revenue_trans.merge(transactions, how='left', left_on='payment_key', right_on='payment_key').groupby('trans_type').sum()
print(quantity_trans.head(10))
print(revenue_trans.head())
```

```python
# product correlation vs store division

rev_by_division = facts[['store_key', 'item_key', 'total_price']].merge(stores, how='left', left_on='store_key', right_on='store_key')
rev_by_division = rev_by_division.merge(items[['item_key', 'desc']], how='left', left_on='item_key', right_on='item_key')
table = pd.pivot_table(rev_by_division, values = 'total_price', index = ['desc'], columns = ['division'], aggfunc="sum")
table = table.apply(lambda x: x/sum(x), axis = 0)
table
```

```python
sns.heatmap(table)
```

```python
# same thing for transactions per store

trans_by_division = facts[['store_key', 'payment_key', 'total_price']].merge(stores, how='left', left_on='store_key', right_on='store_key')
trans_by_division = trans_by_division.merge(transactions[['payment_key', 'trans_type']], how='left', left_on='payment_key', right_on='payment_key')
table = pd.pivot_table(trans_by_division, values = 'total_price', index = ['trans_type'], columns = ['division'], aggfunc="sum")
table = table.apply(lambda x: x/sum(x), axis = 0)
table
```

```python
sns.heatmap(table)
```
